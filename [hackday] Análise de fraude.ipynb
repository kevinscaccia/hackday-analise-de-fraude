# Nome e EMAIL
## Criar score de confiança entre nome e email = Similaridade
Utilizar Técnica de JaroWinkler

- Limpeza de espaços em branco entre os nomes

- Limpeza de pontos no email
 - Criar um peso mais significativo para direcionar a confiança

```python
# pip install fuzzywuzzy
from fuzzywuzzy import fuzz
```

```python
# !pip install jaro-winkler
import jaro
from unidecode import unidecode
```

```python
# Precisamos padronizar os nomes e emails
# colocando nomes e sobrenomes em minusculos e retirando os acentos dos nomes para comparar com email
data['nome1'] = [unidecode(i.lower()) for i in data['nome']]
juntarNomes = [''.join(unidecode(i.lower()).split(' ')) for i in data['nome']] # precisamos ver como deixar nossa base com nome completo ou não
data['sobrenome1'] = [unidecode(i.lower()) for i in data['sobrenome']]


# e prefixos dos emails, separados dos domínios
data['prefixo_email'] = [i.split('@')[0] for i in data['email']]

data[['nome', 'nome1', 'sobrenome', 'sobrenome1', 'email', 'prefixo_email']].head(10)
```

```python
def comparaTecnicas(prefixo, nomeCompleto, tecnica):
    if tecnica == 'token_set_ratio':
        return fuzz.token_set_ratio(nomeCompleto, prefixo)
    elif tecnica == 'jaro_winkler_metric':
        return jaro.jaro_winkler_metric(nomeCompleto, prefixo)
```

# Incluir no dataset variações nos emails

```python
for i in range(0, 10):
    prefixo = ''.join(data['prefixo_email'].iloc[i])
    nomeCompleto = data['nome1'].iloc[i] + data['sobrenome1'].iloc[i]

    print(nomeCompleto, '-',prefixo)
    print('|', f'token_set_ratio: {comparaTecnicas(prefixo, nomeCompleto, "token_set_ratio")}', 
          '|', f'jaro_winkler: {comparaTecnicas(prefixo, nomeCompleto, "jaro_winkler_metric")}', '|')
```
